```python
#ex1

import numpy as np
import matplotlib.pyplot as plt
import scipy.io.wavfile as wav
import scipy.signal
import sounddevice as sd

t = np.linspace(0, 5)
f = 1000
x = np.sin(2 * np.pi * f * t + np.pi / 3)
plt.suptitle("Sine wave")
plt.plot(t, x)
plt.grid(True)
plt.axhline(y=0, color='black')
plt.show()


x2 = np.cos(2 * np.pi * f * t - np.pi / 6)
plt.suptitle("Cosine wave")
plt.plot(t, x2)
plt.grid(True)
plt.axhline(y=0, color='black')
plt.show()
```

```python
#ex2

t = np.linspace(0, 5)
f = 1000
ls = [0, np.pi / 3, np.pi / 4, 3 * np.pi / 2]

for i in ls:
    x = np.sin(2 * np.pi * f * t + i)
    plt.plot(t, x)

plt.suptitle("Sine waves with phase shift")
plt.grid(True)
plt.axhline(y=0, color='black')
plt.show()
```
