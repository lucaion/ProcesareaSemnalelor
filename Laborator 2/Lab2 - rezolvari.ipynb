```python
#ex1

import numpy as np
import matplotlib.pyplot as plt
import scipy.io.wavfile as wav
import scipy.signal
import sounddevice as sd

t = np.linspace(0, 5)
f = 1000
phi = 0
x = np.sin(2 * np.pi * f * t + phi)
x2 = np.cos(2 * np.pi * f * t + (phi - np.pi / 2))

fig, ax = plt.subplots(2, dpi=120)
fig.suptitle("Ex 1")
ax[0].set_title('Sine wave')
ax[0].plot(t, x)
ax[1].set_title('Shifted cosine wave')
ax[1].plot(t, x2)
for ax in ax.flat:
    ax.axhline(y=0, color='black')
    ax.set(xlabel="Time", ylabel="Amplitude")
    ax.grid(True)

plt.subplots_adjust(hspace=1)
plt.show()
```

```python
#ex2

t = np.linspace(0, 5)
f = 1000
ls = [0, np.pi / 3, np.pi / 4, 3 * np.pi / 2]

for i in ls:
    x = np.sin(2 * np.pi * f * t + i)
    plt.plot(t, x)

plt.title('Ex 2')
plt.grid(True)
plt.axhline(y=0, color='black')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()
```

```python
#ex4

t = np.linspace(0, 5, 100)
f = 4000
x = np.sin(2 * np.pi * f * t)
y = scipy.signal.sawtooth(2 * np.pi * 1 * t)

fig, ax = plt.subplots(3, dpi=120)
fig.suptitle("Ex 4")
ax[0].set_title('Sine wave')
ax[0].plot(t, x)
ax[1].set_title('Sawtooth wave')
ax[1].plot(t, y)
ax[2].set_title('Sine + sawtooth wave')
ax[2].plot(t, x + y)

for ax in ax.flat:
    ax.axhline(y=0, color='black')
    ax.set(xlabel="Time", ylabel="Amplitude")
    ax.grid(True)

plt.subplots_adjust(hspace=2)
plt.show()
```

```python
#ex5

t = np.linspace(0, 2, 400)
t2 = np.linspace(0, 2, 800)
f = 1000
f2 = 200

x = np.sin(2 * np.pi * f * t)
y = np.sin(2 * np.pi * f2 * t)
z = np.concatenate((x, y))

fig, ax = plt.subplots(3, dpi=120)
fig.suptitle("Ex 5")
ax[0].set_title('x')
ax[0].plot(t, x)
ax[1].set_title('y')
ax[1].plot(t, y)
ax[2].set_title('z')
ax[2].plot(t2, z)

for ax in ax.flat:
    ax.axhline(y=0, color='black')
    ax.set(xlabel="Time", ylabel="Amplitude")
    ax.grid(True)

plt.subplots_adjust(hspace=2)

plt.show()

fs = 44100
sd.play(z, fs)
```

```python
#ex6

t = np.linspace(0, 6, 800)
fs = 3200

x = np.sin(2 * np.pi * fs / 2 * t)
y = np.sin(2 * np.pi * fs / 4 * t)
z = np.sin(2 * np.pi * 0 * t)

fig, ax = plt.subplots(3, dpi=120)
fig.suptitle("Ex 6")
ax[0].set_title('x')
ax[0].plot(t, x)
ax[1].set_title('y')
ax[1].plot(t, y)
ax[2].set_title('z')
ax[2].plot(t, z)

for ax in ax.flat:
    ax.set(xlabel="Time", ylabel="Amplitude")
    ax.grid(True)

plt.subplots_adjust(hspace=2)

plt.show()
```

```python
#ex7

t = np.linspace(0, 3, 40)
td = np.linspace(0, 3, 10)
f = 1000

x = np.sin(2 * np.pi * f * t)
x2 = x[0::4]
x3 = x[1::4]

fig, ax = plt.subplots(3, dpi=120)
fig.suptitle("Ex 7")
ax[0].set_title('x')
ax[0].plot(t, x)
ax[1].set_title('x2')
ax[1].plot(td, x2)
ax[2].set_title('x3')
ax[2].plot(td, x3)

for ax in ax.flat:
    ax.set(xlabel="Time", ylabel="Amplitude")
    ax.grid(True)

plt.subplots_adjust(hspace=2)

plt.show()
```

```python
#ex8

t = np.linspace(-np.pi / 4, np.pi / 4)
fig, ax = plt.subplots(2, dpi=120)
fig.suptitle("Ex 8")
ax[0].set_title('sin(t)')
ax[0].plot(t, np.sin(t))
ax[1].set_title('t')
ax[1].plot(t, t)

for ax in ax.flat:
    ax.set(xlabel="Time", ylabel="Amplitude")
    ax.grid(True)

plt.subplots_adjust(hspace=1)
plt.show()
```
