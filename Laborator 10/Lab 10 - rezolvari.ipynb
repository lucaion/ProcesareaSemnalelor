```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ex 1

mean_1d = 2
variance_1d = 3
std_dev_1d = np.sqrt(variance_1d)

samples_1d = np.random.normal(mean_1d, std_dev_1d, 1000)

plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.hist(samples_1d, bins=30, density=True, alpha=0.7, color='blue')
plt.title('Distribuie Gaussiana Unidimensionala')

mean_2d = np.array([2, 3])
covariance_matrix_2d = np.array([[3, 1], [1, 2]])

eigenvalues, eigenvectors = np.linalg.eigh(covariance_matrix_2d)

n_samples = 1000
random_normal_samples = np.random.standard_normal(size=(2, n_samples))
scaled_samples_2d = eigenvectors @ np.diag(np.sqrt(eigenvalues)) @ random_normal_samples + mean_2d[:, np.newaxis]

plt.subplot(1, 2, 2)
plt.scatter(scaled_samples_2d[0], scaled_samples_2d[1], alpha=0.5, color='red')
plt.title('Distribuie Gaussiana Bidimensionala')

plt.tight_layout()
plt.show()
```

```python
# ex2

def k(x, y):
  return np.dot(x, y)

x = np.linspace(-1, 1, 100)

C = np.zeros((100, 100))
for i in range(100):
  for j in range(100):
    C[i, j] = k(x[i], x[j])

z = np.random.multivariate_normal(np.zeros(100), C)

plt.plot(x, z)
plt.title("Proces Gaussian - Liniar")
plt.show()
```

```python
def k(s, t):
  return np.minimum(s, t)

t = np.linspace(-1, 1, 100)

C = np.zeros((100, 100))
for i in range(100):
  for j in range(100):
    C[i, j] = k(t[i], t[j])

z = np.random.multivariate_normal(np.zeros(100), C)

plt.plot(t, z)
plt.title('Proces Gaussian - Miscare Browniana')
plt.show()
```

```python
alpha = 500

def k(x, y):
  return np.exp(-alpha * (np.linalg.norm(x-y) ** 2))

t = np.linspace(-1, 1, 100)

C = np.zeros((100, 100))
for i in range(100):
  for j in range(100):
    C[i, j] = k(t[i], t[j])

z = np.random.multivariate_normal(np.zeros(100), C)

plt.plot(t, z)
plt.title('Proces Gaussian - Exponentiala Patrata')
plt.show()
```

```python
alpha = 5

def k(s, t):
  return np.exp(-alpha * np.abs(s - t))

t = np.linspace(-1, 1, 100)

C = np.zeros((100, 100))
for i in range(100):
  for j in range(100):
    C[i, j] = k(t[i], t[j])

z = np.random.multivariate_normal(np.zeros(100), C)

plt.plot(t, z)
plt.title('Proces Gaussian - Ornstein-Uhlenbeck')
plt.show()
```

```python
alpha = 2
beta = 1

def k(x, y):
  return np.exp(-alpha * np.sin(beta * np.pi * (x - y)) ** 2)

x = np.linspace(-1, 1, 100)

C = np.zeros((100, 100))
for i in range(100):
  for j in range(100):
    C[i, j] = k(x[i], x[j])

z = np.random.multivariate_normal(np.zeros(100), C)

# Afisarea rezultatelor
plt.plot(x, z)
plt.title('Proces Gaussian - Periodic')
plt.show()
```

```python
alpha = 50

def k(x, y):
  return np.exp(-alpha * np.minimum(np.abs(x - y), np.abs(x + y)) ** 2)

x = np.linspace(-1, 1, 100)

C = np.zeros((100, 100))
for i in range(100):
  for j in range(100):
    C[i, j] = k(x[i], x[j])

z = np.random.multivariate_normal(np.zeros(100), C)
# Afisarea rezultatelor
plt.plot(x, z)
plt.title('Proces Gaussian - Simetric')
plt.show()
```
