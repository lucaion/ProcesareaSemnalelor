```python
import numpy as np
import matplotlib.pyplot as plt

#ex1
#a
duration = 0.03
t = np.arange(0, duration, 0.0005)


def signal_x(x):
    return np.cos(520 * np.pi * x + np.pi / 3)


def signal_y(x):
    return np.cos(280 * np.pi * x - np.pi / 3)


def signal_z(x):
    return np.cos(120 * np.pi * x + np.pi / 3)


value_x = signal_x(t)
value_y = signal_y(t)
value_z = signal_z(t)

#b
fig, ax = plt.subplots(3, dpi=150)
fig.suptitle("Ex 1b")
ax[0].set_title('x(t)')
ax[0].plot(t, value_x)
ax[1].set_title('y(t)')
ax[1].plot(t, value_y)
ax[2].set_title('z(t)')
ax[2].plot(t, value_z)

for ax in ax.flat:
    ax.axhline(y=0, color='black')
    ax.set(xlabel="Time", ylabel="Amplitude")
    ax.grid(True)

plt.subplots_adjust(hspace=2)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.show()
```

```python
#c
sampling_rate = 200
num_samples = int(sampling_rate * duration)
samples = np.linspace(0, duration, num_samples)

samples_x = signal_x(samples)
samples_y = signal_y(samples)
samples_z = signal_z(samples)

fig, ax2 = plt.subplots(3, dpi=150)

fig.suptitle("Ex 1c")

ax2[0].set_title('x[t]')
ax2[0].plot(t, value_x)
ax2[0].stem(samples, samples_x, linefmt='orange')
ax2[1].set_title('y[t]')
ax2[1].plot(t, value_y)
ax2[1].stem(samples, samples_y, linefmt='orange')
ax2[2].set_title('z[t]')
ax2[2].plot(t, value_z)
ax2[2].stem(samples, samples_z, linefmt='orange')

plt.subplots_adjust(hspace=2)

for subplot in ax2:
    subplot.grid(True)
    subplot.set(xlabel="Time", ylabel="Amplitude")
    subplot.axhline(y=0, color='black')

plt.show()
```

```python
#ex2
#a
t = np.linspace(0, 0.03, 1600)
f = 400
x = np.sin(2 * np.pi * f * t)
plt.suptitle('Sine wave 400 Hz')
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.plot(t, x)
plt.grid(True)
plt.axhline(y=0, color='black')
plt.show()
```

```python
#b
t = np.linspace(0, 3)
f = 800
x = np.sin(2 * np.pi * f * t)
plt.suptitle("Sine wave 800 Hz")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.plot(t, x)
plt.grid(True)
plt.axhline(y=0, color='black')
plt.show()
```

```python
#c
t = np.linspace(0, 5, 5 * 240)
x = np.mod(t, 2)
plt.suptitle('Sawtooth wave 240 Hz')
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.plot(t, x)
plt.grid(True)
plt.axhline(y=0, color='black')
plt.show()
```

```python
#d
t = np.linspace(0, 5, 5 * 300)
x = np.sign(np.sin(2 * np.pi * t))
plt.suptitle('Square wave 300 Hz')
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.plot(t,x)
plt.grid(True)
plt.axhline(y=0, color='black')
plt.show()
```
